```python
# Parse year from filename
import re

def extract_year_from_filename(path: Path) -> int:
    """
    Extracts the first 4-digit year from the filename.
    """
    m = re.search(r"(20[0-2][0-9])", path.name)
    return int(m.group(1)) if m else None

# Find summary files
summary_files = sorted(DATA_RAW_PATH.glob("*Summary*.csv"))
summary_files
```

```python
# Load summary files and add year column
summary_dfs = []

for f in summary_files:
    year = extract_year_from_filename(f)
    print(f"Loading", f.name, "year =", year)
    df = pd.read_csv(f)
    df["year"] = year
    summary_dfs.append(df)

summary_all = pd.concat(summary_dfs, ignore_index=True)
summary_all.shape
```
