```python
# Import and load raw data
import pandas as pd
from pathlib import Path

CLEAN_PATH = Path("../data/clean")
RAW_PATH = Path("../data/raw")

# Load the raw combined summary created in Notebook 1
summary = pd.read_csv(CLEAN_PATH / "summary_raw_2020_2022.csv")
summary.head()

```

```python
# Remove data columns that are not useful for analysis
cols_to_drop = [
    "MedicalDirector",
    "Address",
    "ZipCode",
    "Phone",
    "DisplayOrder",
    "TopicId",
    "SubTopicId",
    "QuestionId",
    "BreakoutCategoryId",
    "BreakoutId",
    "Data_Value_Footnote_Symbol",
    "Data_Value_Footnote"
]

summary_clean = summary.drop(columns=cols_to_drop, errors="ignore")
summary_clean.head()
```

```python
# Drop rows without Data_Value
summary_clean = summary_clean.dropna(subset=["Data_Value"])
summary_clean.shape
```

```python
# Standardize column names
summary_clean.columns = (
    summary_clean.columns
    .str.lower()
    .str.replace(" ", "_")
    .str.replace("__", "_")
)
summary_clean.head()
```

```python
summary_clean.to_csv("../data/clean/summary_clean.csv", index=False)
```
